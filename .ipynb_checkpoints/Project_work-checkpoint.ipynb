```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# importing the red_wine.csv dataset into the working notebook environment
df = pd.read_csv('red_wine.csv')
```

```python
# Previewing the first 10 rows using the head method
df.head(10)
```

```python
# Previewing the last 20 rows using the tail method
df.tail(20)
```

```python
# Using the info method to check basic information on the dataset
df.info()
```

```python
#Using the describe method to obtain statistical information on the dataset
df.describe()
```

## A box plot of wine quality against alcohol content 

```python
plt.figure()
sns.boxplot(x='quality',y='alcohol',data = df,palette= 'dark:red', hue= 'quality')
plt.xlabel('Quality (%)')
plt.ylabel('Alcohol')
plt.title('Wine Quality vs Alcohol Content')
plt.show()
```

## Pairplot of Key Features
* This pairplot displays scatterplots of all possible pairs of variables in the dataset. By examining the points in the scatterplots, the trends, patterns, clusters, or outliers, providing a qualitative understanding of the data can be identified

```python
sns.pairplot(df[['fixed acidity', 'volatile acidity', 'residual sugar', 'total sulfur dioxide', 'alcohol', 'quality']])
plt.suptitle("Pairplot of Key Features", y=1.03)
plt.show()
```

## Distribution of wine quality scores

```python
plt.figure (figsize=(6,5))
sns.displot(df['quality'],kde=False,bins=6)
plt.title('Distribution of Wine Quality Scores')
plt.xlabel('Quality Rating')
plt.ylabel('Count')
plt.show()
```

## Correlation heatmap
* The intensity of color in the correlation heatmap reflects the strength of the correlation. A dark color implies a strong correlation, while a lighter shade suggests a weaker relationship.

```python
plt.figure(figsize=(12,9))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
